```python
import pandas as pd
import numpy as np
import zipfile
import matplotlib.pyplot as plt
```

```python
zip_file_path = 'train.csv.zip'

# Open the ZIP file and list its contents
with zipfile.ZipFile(zip_file_path, 'r') as z:
    # List all files in the ZIP archive
    file_list = z.namelist()
    print("Files in ZIP archive:", file_list)
    
    # Extract the specific CSV file
    csv_file_name = 'train.csv'
    with z.open(csv_file_name) as f:
        train_pd = pd.read_csv(f)

train_pd.head()
n = 199424
```

```python
y_np = train_pd.V1.values[:n]
x_np = train_pd.loc[:,train_pd.columns != "V1"].values[:n, :]
y_np = y_np - 1
```

### Let us visualize some of them

```python
plt.plot(x_np[0])
plt.title('46 days of one piece of land')
plt.xlabel('Days')
plt.ylabel('Scale of grey of the image')
plt.grid(True)
plt.show()
```
